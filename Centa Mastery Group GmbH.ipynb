## PROFIT / LOSS Calculator

### OPTION 1

```python
days = [1,2,3,4,5]
prices = [50,50,55,65,45]
shares = 10
def prof_loss(days, price, share):
    df = pd.DataFrame({'Days':days, 'Prices': prices, 'Shares': shares})
    df['Costs'] = df.Prices * df.Shares
    
    prof_loss = []
    for i in range(len(df)):
        prof_loss.append(df.Costs.iloc[i]-df.Costs.iloc[0])
    df['Profit/Loss'] = prof_loss
    return df
```

```python
prof_loss(days,prices,shares)
```
